## Importing Libraries
TensorFlow and Keras libraries were imported to build and train the neural network model. These libraries provide functions for loading datasets, creating neural network layers, training models, and evaluating prediction accuracy.


```python
import tensorflow as tf
from tensorflow import keras
```

## Load the dataset
The Fashion MNIST dataset was loaded using TensorFlow Keras datasets. This dataset contains grayscale images of different clothing items used for image classification tasks. The dataset was divided into training data and testing data. The image values were normalized by dividing by 255.0 so the neural network could process the data more efficiently during training.


```python
from tensorflow import keras
fashion_mnist = keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

## Build the Neural Network Model
The neural network model was created using the Sequential method in Keras. The model includes a Flatten layer, one hidden Dense layer with ReLU activation, and an output layer for image classification. These layers help the model learn image patterns from the training dataset.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10)
])
```

## Compiling the Model
The model was compiled before training. The Adam optimizer, loss function, and accuracy metric were selected to prepare the model for learning and evaluation.


```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

## Training the Neural Network
The model was trained using the training dataset for five epochs. During training, the neural network learned patterns from the image data and improved its prediction performance.

```python
model.fit(x_train, y_train, epochs=5)
```

## Evaluating the Model Accuracy
The trained model was evaluated using the testing dataset. This step measures how well the model performs on new images that were not used during training. The test accuracy provides an estimate of the model's overall classification performance.

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print("Test accuracy:", test_acc)
```

## Creating the Probability Model
The trained model gives raw prediction scores for each image class. A Softmax layer was added to convert these scores into probability values. These probabilities show how confident the model is for each possible image category.

```python
probability_model = tf.keras.Sequential([
    model,
    tf.keras.layers.Softmax()
])
probability_model(x_test[:5])
```
